```python
import pandas as pd
fnames = ['Working Conditions', 'Relationship with Colleagues', 'Job Satisfaction', 'Company Policy', 'Rewards and Awards', 'Workload and Support']
f_dict = dict()

for i in range(0, 6):
    f_dict[fnames[i]] = data.iloc[:, 4 * i + 1:4 * i + 5].sum(axis=1)

fac_data = pd.DataFrame(f_dict)
print(fac_data.head())
```

```python
fac_data['Score'] = fac_data.iloc[:, 0:6].sum(axis=1)
mean_score = fac_data['Score'].mean()
import numpy as np
fac_data['Satisfaction'] = np.where(fac_data['Score'] >= mean_score, 'High', 'Low')
fac_data.head(10)
```

```python
count=fac_data['Satisfaction'].value_counts()
count
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have your DataFrame fac_data

# Calculate the correlation matrix
cor = fac_data.iloc[:, 0:6].corr()

# Create a clustered heatmap without using seaborn's clustermap
plt.figure(figsize=(8, 6))
sns.heatmap(cor, cmap='coolwarm', annot=True)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Set the title
plt.title("Clustered Correlation Matrix Heatmap")

# Display the heatmap
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster import hierarchy
from sklearn.preprocessing import StandardScaler

linkage_matrix = hierarchy.linkage(cor, method='ward')

# Create a dendrogram
dendrogram = hierarchy.dendrogram(linkage_matrix, labels=cor.columns, orientation='top')
plt.xticks(rotation=90)
plt.title("Dendrogram of Correlation Matrix")
plt.show()
```
